```python
#from https://github.com/savarin/pyconuk-introtutorial

# load training data
import pandas as pd
import numpy as np

df = pd.read_csv('./train.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
# drop irrelevant columns(by eyes)
df = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
# Check and convert categorical Sex and Embarked into numerical ones.
df['Sex'].unique()
```

```python
df['Gender'] = df['Sex'].map({'female': 0, 'male': 1}).astype(int)
```

```python
df['Embarked'].unique()
```

```python
df['Port']=df['Embarked'].map({'C': 1, 'S':2, 'Q':3}).astype(int)
```

```python
df = df.drop(['Sex', 'Embarked'], axis=1)
```

```python
cols = df.columns.tolist()
print(cols)
```

```python
cols = [cols[1]] + cols[0:1] + cols[2:]
df = df[cols]
df.head()
```

```python
df.info()
```

```python
train_data = df.values # convert to numerical array
```

```python
train_data
```

```python
# random forest algorithm
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100)

#We use the processed training data to 'train' (or 'fit') our model. The column Survived will be our second input, and the set of other features (with the column PassengerId omitted) as the first.
model = model.fit(train_data[0:, 2:], train_data[0:,0])

```

```python
train_data[0:, 0].size
```

```python
train_data.shape
```

```python
df_test = pd.read_csv('./test.csv')
```

```python
df_test.head(10)
```

```python
df_test = df_test.drop(['Name','Ticket', 'Cabin'], axis=1)

df_test = df_test.dropna()

df_test['Gender'] = df_test['Sex'].map({'female': 0, 'male':1})
df_test['Port']= df_test['Embarked'].map({'C':1, 'S':2, 'Q':3})

df_test = df_test.drop(['Sex', 'Embarked'], axis=1)

test_data = df_test.values
```

```python
output = model.predict(test_data[:, 1:])
```

```python
# Pandas - Preparing for submission

result = np.c_[test_data[:,0].astype(int), output.astype(int)]
df_result = pd.DataFrame(result[:, 0:2], columns=['PassengerId', 'Survived'])
```

```python
df_result.head(10)
```

```python
df_result.to_csv('./titanic_1-0.csv', index=False)
```
