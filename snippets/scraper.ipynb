##  Simple html element selector

```python
from bs4 import BeautifulSoup # python3 and python2, installed with pip install beautifulsoup4, 
 # tutorial is https://www.crummy.com/software/BeautifulSoup/bs4/doc/

html = "<html><body><div class='container'>Hello world</div><div><i>Hello world</i></div></body></html>"#the HTML code you've written above
doc = BeautifulSoup(html, "html5lib")

print(doc.body.find('div', attrs={'class':'container'}).text)
print(doc.find(text="Hello world"))
doc.i
doc.div
```

```python
import urllib.request

link = "http://www.163567.com/history/2017.html"
response = urllib.request.urlopen(link)
r = response.read()
doc2 = BeautifulSoup(r, "html5lib") # 会自动处理编码问题
```

```python
table = doc2.body.find('table', attrs={"id":"main"})
rows = table.find_all('tr', attrs={"class":"infolist"})
rows
```
