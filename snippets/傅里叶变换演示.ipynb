```python
import matplotlib as mp
from pylab import *
import numpy as np
x = np.linspace(0, 2 * np.pi, 101)
y = np.sin(x)
z = np.cos(x*2)
plot(x,y);grid();title("sin x")
plot(x,z);grid();title("sin x*2")
show()
```

### 傅里叶变换演示

**见文章[如果看了此文你还不懂傅里叶变换，那就过来掐死我吧【完整版】](http://blog.jobbole.com/70549/)**

```python
y1 = 4*sin(x)/np.pi
plot(x,y1);grid();title("y1:4sin(x)/pi");show();

y3 = 4*sin(3*x)/(3*np.pi)
plot(x,y3);grid();title("y3:4sin(3x)/3*pi");

y5 = 4*sin(5*x)/(5*np.pi)
plot(x,y5);grid();title("y5:4sin(5x)/5pi");

y7 = 4*sin(7*x)/(7*np.pi)
plot(x,y7);grid();title("y7:4sin(7x)/7pi");show();


plot(x,y1 + y3);grid();title("y1 + y3");show();
plot(x,y1 + y3 + y5);grid();title("y1 + y3 + y5");show();

plot(x,y1 + y3 + y5 + y7);grid();title("y1 + y3 + y5 + y7");show();
```
