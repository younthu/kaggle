```python
# plotting cross validated predictions, 
# http://scikit-learn.org/stable/auto_examples/plot_cv_predict.html#sphx-glr-auto-examples-plot-cv-predict-py

# 交叉验证方法是验证方法，不是预测方法，随机把数据拆成训练集和测试集，然后计算准确值。如此轮换几次求平均值

# 练习目的: 
# 1. 用cross validation 来验证不同的模型效果
# 2. 理解cross validated predictions值的意义
# 3. boston['DESCR']打印波士顿房价数据的描述

from sklearn import datasets
from sklearn.model_selection import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt

lr = linear_model.LinearRegression()
boston = datasets.load_boston()
y = boston.target

def plot_with_cv(cv):
    # cross_val_predict returns an array of the same size as 'y' where each entry
    # is a prediction obtained by cross validataion:
    predicated = cross_val_predict(lr, boston.data, y, cv=cv) # cv, 交叉迭代多少次

    fig, ax = plt.subplots()
    ax.scatter(y, predicated, edgecolors=(0,0,0), alpha=0.3)
    ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)#，画直线， k--,  黑色虚线，lw=4,line width=4
    ax.set_xlabel('Measured')
    ax.set_ylabel('Predicted')
    plt.show()
```

```python
print(y.shape)
print(y.min(),y.max())
plot_with_cv(10)
```

```python
plot_with_cv(100)
```
