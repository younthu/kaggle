```python
%matplotlib inline
import pylab as pl # matplotlib的一个模块
import numpy as np
import scipy.stats as stats

#二项分布
n,p=50, 0.2
k = np.arange(n+1)
x = stats.binom.pmf(k,n,p)

pl.stem(k,x)
pl.show()
```

```python
#泊松分布
n, mu = 50,10
k = np.arange(n+1)
x = stats.poisson.pmf(k, mu)

pl.stem(k, x)
pl.show()
```

```python
#几何分布

n, p = 50, 0.2
k = np.arange(n+1)
x = stats.geom.pmf(k, p)

pl.stem(k, x)
pl.show()
```

```python
# 正态分布（高斯分布）

x = np.linspace(-5, 5, 100)
f = stats.norm.pdf(x)

pl.plot(x, f)
pl.show()
```

```python
x = np.linspace(0, 10, 100)
f = stats.expon.pdf(x)

pl.plot(x, f)
pl.show()
```
